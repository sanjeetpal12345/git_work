```python
from collections import Counter

# Define file paths
file_path1 = 'C:\\Users\\hp.DESKTOP-MTCIJR8\\Videos\\dfccil\\work\\ATCbuybackprinter1.txt'
file_path2 = 'C:\\Users\\hp.DESKTOP-MTCIJR8\\Videos\\dfccil\\work\\printerbuyback2.txt'

# Function to extract serial numbers from a file
def extract_serial_numbers(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read().lower()
        # Split the content into words and filter alphanumeric words
        words = content.split()
        serial_numbers = [word for word in words if word.isalnum()]  # List instead of set for counting duplicates
        return serial_numbers

# Function to get duplicates from a list of serial numbers
def get_duplicates(serial_numbers):
    counter = Counter(serial_numbers)
    duplicates = {serial: count for serial, count in counter.items() if count > 1}
    return duplicates

# Extract serial numbers from both files
serial_numbers1 = extract_serial_numbers(file_path1)
serial_numbers2 = extract_serial_numbers(file_path2)

# Find duplicates in each file
duplicates_in_file1 = get_duplicates(serial_numbers1)
duplicates_in_file2 = get_duplicates(serial_numbers2)

# Find common serial numbers
common_serial_numbers = set(serial_numbers1).intersection(set(serial_numbers2))

# Find unique serial numbers in each file
unique_in_file1 = set(serial_numbers1) - set(serial_numbers2)
unique_in_file2 = set(serial_numbers2) - set(serial_numbers1)

# Print the common serial numbers
print("Common serial numbers:", common_serial_numbers)

# Print unique serial numbers
print("\nUnique serial numbers in file 1:")
print(unique_in_file1)

print("\nUnique serial numbers in file 2:")
print(unique_in_file2)

# Print duplicates
print("\nDuplicates in file 1:")
print(duplicates_in_file1)

print("\nDuplicates in file 2:")
print(duplicates_in_file2)

# Count and print the number of matches and unique serial numbers
print("\nTotal number of common serial numbers:", len(common_serial_numbers))
print("Total number of unique serial numbers in file 1:", len(unique_in_file1))
print("Total number of unique serial numbers in file 2:", len(unique_in_file2))

# Count and print the number of duplicates
print("\nTotal number of duplicates in file 1:", len(duplicates_in_file1))
print("Total number of duplicates in file 2:", len(duplicates_in_file2))
```
